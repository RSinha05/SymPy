```python
from sympy import *
x = Symbol('x')
```

```python
from sympy.plotting import *
plot_backends
```

```python
plot_backends['default'] = plot_backends['ipythonD3']
```

```python
plot_backends
```

```python
plot_parametric((sin(x)*x, cos(x)*x), 
                (sin(x)*x, -cos(x)*x),
                (-sin(x)*x, -cos(x)*x),
                (-sin(x)*x, cos(x)*x),
                (sin(x)*x, cos(x)*x)[::-1], 
                (sin(x)*x, -cos(x)*x)[::-1],
                (-sin(x)*x, -cos(x)*x)[::-1],
                (-sin(x)*x, cos(x)*x)[::-1],

                (x,0,2*pi));
```

```python
plot(sqrt(abs(x)), -sqrt(abs(x)),
     sin(x), -sin(x),
     cos(x), -cos(x),
     sin(x)/x, -sin(x)/x,
     tanh(x), -tanh(x),

     (x,-2*pi,+2*pi));
```
